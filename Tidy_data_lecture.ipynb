```python
# billboard
import env


def get_connection(db, user=env.user, host=env.host, password=env.password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'

sql_query = """
                SELECT * 
From billboard;
                """

billboard = pd.read_sql(sql_query, get_connection('tidy_data'))
```

### Variable are stored in both rows and columns

```python
#weather
sql_query = """
                SELECT * 
From weather;
                """

weather = pd.read_sql(sql_query, get_connection('tidy_data'))
```

```python
# this is wide, tmax, day across the top
# you cant analyze this data set because of the way is structured
# days into rows and , elements into columns, it is also a measurement and measurements go into columns
weather.head()
```

```python
# get day columns into rows
weather_long = weather.melt(id_vars = ['id', 'year','month','element'],
            var_name = 'days',
            value_name = 'temp')
weather_long
```

```python
# get rid of nan by pivot element columns so where there is a nan we can get rid of these values
weather_long.days = weather_long.days.str[1:].astype(int)
weather_long
```

```python

weather_long.pivot_table(index=['id','year','month','days'],
                        columns = 'element', # this columns pivots
                        values ='temp').reset_index() # this is where it matches, reset_index for a better table
```

```python
#weather
sql_query = """
                SELECT * 
From sales;
                """

sales = pd.read_sql(sql_query, get_connection('tidy_data'))
```

```python
sales
```

```python
# columns to a row is a melt
sales_melt = sales.melt(id_vars='Product')
```

```python
#object column with string
sales_melt.variable
```

```python
# use split to seperate the names and set expand = True to get two columns
sales_melt.variable.str.split(' ',expand = True)
```

```python
# join back to melt
sales_melt[['year','measure']]= sales_melt.variable.str.split(' ',expand = True)
```

```python
sales_melt
```

```python
# multy indexed Product and Measure is an index
sales_melt.pivot_table(index=['Product','year'], #product and year must match, keep as columns
                      columns = 'measure', #change measure to columns create columns of its values: ppu, sales
                      values ='value') # values of the columns section of previous line
```

```python
# got rid of variable because i didnt mention it. 
sales_tidy = sales_melt.pivot_table(index=['Product','year'], #product and year must match, keep as columns
                      columns = 'measure', #change measure to columns create columns of its values: ppu, sales
                      values ='value').reset_index() # values of the columns section of previous line

sales_tidy
```

```python
# to get rid of the index name measure
sales_tidy = sales_tidy.rename_axis(None, axis = 1)
sales_tidy
```
